```python
import pandas as pd
import numpy as np
```

## Refinement 3 to 4

```python
df = pd.read_excel('/content/WOT_internal_refined_3_condensed_spans.xlsx')
```

```python
df['answer_start'] = df['Annotated'].apply(lambda x: [i for i,val in enumerate(x) if val=='<'])
df['answer_end'] = df['Annotated'].apply(lambda x: [i for i,val in enumerate(x) if val=='>'])
df['answer_extr'] = df.apply(lambda row: [row['Annotated'][row['answer_start'][i]+1:row['answer_end'][i]] for i, val in enumerate(row['answer_start'])], axis=1)
df['number_answer_elements'] = df['Annotated'].apply(lambda x: len([i for i,val in enumerate(x) if val=='<']))

df['Context'] = df['Annotated']
df['Context'] = df['Context'].apply(lambda x: str(x).replace('>', ''))
df['Context'] = df['Context'].apply(lambda x: str(x).replace('<', ''))

df.to_excel('/content/WOT_internal_refined_4.xlsx')
```

## Refinement 5 to 6 rework Taxonomy

```python
df = pd.read_excel('../WOT_internal_refined_5.xlsx')
df['label'] = df['label'].apply(lambda x: x.split(' + '))
df['is_factoid'] = df['label'].apply(lambda x: 'factoid' in x)
df['is_confirmation'] = df['label'].apply(lambda x: 'confirmation' in x)
df['is_complex'] = df['label'].apply(lambda x: 'complex' in x)
df['is_causal'] = df['label'].apply(lambda x: 'causal' in x)
df['is_listing'] = df['label'].apply(lambda x: 'listing' in x)
df['is_history'] = df['label'].apply(lambda x: 'history' in x)
df['is_navigation'] = df['label'].apply(lambda x: 'navigation' in x)
df['is_unanswerable'] = df['label'].apply(lambda x: 'unanswerable' in x)

df['answer_extr'] = df['answer_extr'].apply(lambda x: x.strip('[]').strip("'"))

df['answer_start'] = df['answer_start'].fillna('[]').astype('str')
df['answer_start'] = df['answer_start'].apply(lambda x: x.strip('[]'))
df['answer_start'] = df['answer_start'].apply(lambda x: np.nan if x=='' else x)
df['answer_start'] = df['answer_start'].astype('float')

df['answer_end'] = df['answer_end'].fillna('[]').astype('str')
df['answer_end'] = df['answer_end'].apply(lambda x: x.strip('[]'))
df['answer_end'] = df['answer_end'].apply(lambda x: np.nan if x=='' else x)
df['answer_end'] = df['answer_end'].astype('float')

df['answers_extr'] = df.apply(lambda row: {'text': row['answer_extr'], 'answer_start': row['answer_start']} , axis=1)

df = df.rename(columns={'label':'labels', 'answer': 'answer_orig_wot'})
```

```python
df.to_excel('../WOT_internal_refined_6.xlsx')
```
