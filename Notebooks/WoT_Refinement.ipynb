```python
import pandas as pd
```

```python
df = pd.read_excel('/content/WOT_internal_refined_3_condensed_spans.xlsx')
```

```python
df['answer_start'] = df['Annotated'].apply(lambda x: [i for i,val in enumerate(x) if val=='<'])
df['answer_end'] = df['Annotated'].apply(lambda x: [i for i,val in enumerate(x) if val=='>'])
df['answer_extr'] = df.apply(lambda row: [row['Annotated'][row['answer_start'][i]+1:row['answer_end'][i]] for i, val in enumerate(row['answer_start'])], axis=1)
df['number_answer_elements'] = df['Annotated'].apply(lambda x: len([i for i,val in enumerate(x) if val=='<']))

df['Context'] = df['Annotated']
df['Context'] = df['Context'].apply(lambda x: str(x).replace('>', ''))
df['Context'] = df['Context'].apply(lambda x: str(x).replace('<', ''))

df.to_excel('/content/WOT_internal_refined_4.xlsx')
```
